```julia
# •·•·•·•·•·•·•·•·•·••·•·•·•·•·•·•·•·•·••·•·•·•·•·•·•·•·•·••·•·•·•·•·•·•·•·•·••·•·•·•·•·•·•·•·•·••·•·•·•·•·•·• #

i = 1
FILEBRW = SearchDir( PATHBRWs, ".brw" )[ i ];
Variables, FILEPATHS = GetVarsHDF5( FILEBRW );

# •·•·•·•·•·•·•·•·•·••·•·•·•·•·•·•·•·•·••·•·•·•·•·•·•·•·•·••·•·•·•·•·•·•·•·•·••·•·•·•·•·•·•·•·•·••·•·•·•·•·•·• #
```

```julia
using JLD
using Colors
using StatsBase
using Plots
```

```julia
#
PATHS = load( FILEPATHS )[ "PATHS" ];
PATHMAIN = PATHS[ "PATHMAIN" ] = pwd( );
PATHFIGURES = joinpath( PATHMAIN, "Figures" ); mkpath( PATHFIGURES );
# Deteccion de canales activos de forma incial antes de abrir el archivo
nChs = Variables[ "nChs" ];
aux = Variables[ "STD" ];
channels = collect( 1:nChs );
NRecFrames = Variables[ SearchKey( "Frame", Variables )[ 1 ] ];
SamplingRate = Variables[ SearchKey( "Rat", Variables )[ 1 ] ];
Rows = getindex.( aux, 1 ); Columns = getindex.( aux, 2 );
STDs = getindex.( aux, 3 ); Means = getindex.( aux, 4 );
Index = zeros( Int, length( Rows ) );
for j = 1:length( Rows )
    Index[ j ] = ( Rows[ j ] - 1 )*64 + Columns[ j ];
end
limlower = ceil( Int, nChs/8 );
limupper = ceil( Int, nChs/2 );
```

```julia
Z = zeros( Float64, nChs ); Z[ Index ] .= STDs;
cm = reverse( colormap( "RdBu", length( countmap( Z ) ); mid = 0.5, logscale = false ) );
P1 = Zplot( log.( Z ), "W", cm );
P1 = plot!( title = "Evaluacion inicial sin abrir el archivo", titlefontsize = 8 );
savefig( P1, joinpath( PATHFIGURES, "ACDv0.png" ) );
savefig( P1, joinpath( PATHFIGURES, "ACDv0.svg" ) );
```

```julia
#
logT, logt = Thresholding( log.( Z ) );
T, t = Thresholding( Z );
P1 = ThresholdingPlots( logT );
P2 = ThresholdingPlots( T );
logTsum = vec( sum( logT, dims = 1 ) );
Tsum = vec( sum( T, dims = 1 ) );
depuratedlogT = logT[ :, limupper .> logTsum .> limlower ];
depuratedT = T[ :, limupper .> Tsum .> limlower ];
depuratedTchannels = hcat( depuratedT.*channels... );
x01 = Int.( keys( countmap( vec( depuratedTchannels ) ) ) );
y = Int.( values( countmap( vec( depuratedTchannels ) ) ) );
x01 = x01[ y .>= ( size( depuratedT, 2 ) - 1 ) ];
x01 = x01[ x01 .!= 0 ];
plot( )
P3 = Zplot( x01, "0" ); P3 = title!( "thresholding raw std" );
depuratedlogTchannels = hcat( depuratedlogT.*channels... );
x02 = Int.( keys( countmap( vec( depuratedlogTchannels ) ) ) );
y = Int.( values( countmap( vec( depuratedlogTchannels ) ) ) );
x02 = x02[ y .>= ( size( depuratedlogT, 2 ) - 1 ) ];
x02 = x02[ x02 .!= 0 ];
plot( )
P4 = Zplot( x02, "0" ); P4 = title!( "thresholding log std" );
plot( )
P5 = Zplot( intersect( x01, x02 ), "0" ); P5 = title!( "Intersection of the 2" );
plot( )
F1 = plot( P3, P4, P5, cbar = :none, layout = ( 1, 3 ), wsize = ( 900, 400 ), titlefontsize = 9 );
savefig( F1, joinpath( PATHFIGURES, "ACDv0T.png" ) );
savefig( F1, joinpath( PATHFIGURES, "ACDv0T.svg" ) );
#
```

```julia
σ = 0;
MaxDescarted = limlower;
PotentialDiscarded = Index[ STDs .< ( mean( STDs ) - σ*std( STDs ) ) ];
while length( PotentialDiscarded ) > MaxDescarted    
    σ = σ + 0.5
    PotentialDiscarded = Index[ STDs .< ( mean( STDs ) - σ*std( STDs ) ) ];
end
σ = 0;
MaxSelected = limupper/2;
PotentialSelected = Index[ STDs .> ( mean( STDs ) + σ*std( STDs ) ) ];
while length( PotentialSelected ) > MaxSelected
    σ = σ + 0.5
    PotentialSelected = Index[ STDs .> ( mean( STDs ) + σ*std( STDs ) ) ];
end
Z = zeros( Int, nChs );
Z[ PotentialSelected ] .= 10;
Z[ PotentialDiscarded ] .= 20;
F1 = Zplot( Z, "W", cm );
F1 = plot!( cbar = :none, title = "Possible Discarted and Selected Channels", titlefontsize = 8 );
savefig( F1, joinpath( PATHFIGURES, "Discarted_Selected.png" ) );
savefig( F1, joinpath( PATHFIGURES, "Discarted_Selected.svg" ) );
```

```julia
x = setdiff( intersect( intersect( x01, x02 ), PotentialSelected ), PotentialDiscarded );
@time grupos, loose = Get_Groups( x );
P1 = FigureGroups( grupos, [ ], :greys );
aux01 = vcat( grupos[ length.( grupos ) .> 3 ]... );
aux02 = FillingHolesCrux( aux01 );
@time grupos, loose = Get_Groups( aux02 );
aux03 = grupos[ length.( grupos ) .> 5 ];
aux03 = vcat( aux03... );
P2 = Zplot( aux03, "0" );
aux04 = setdiff( aux03, PotentialDiscarded );
P3 = Zplot( aux04, "0" );
P4 = FigureGroups( grupos, [ ], :greys );
F1 = plot( P1, P2, P4, P3, layout = ( 2, 2 ), cbar = :none, wszie = ( 700, 700 ) );
```

```julia
limupper = 0.3 # GB
σ = 2
finalsize = ( ( Variables[ "BRWsizeGB" ] / nChs ) * length( aux04 ) ) / ( σ ) ;
while finalsize > limupper 
    σ = σ + 1
    finalsize = ( ( Variables[ "BRWsizeGB" ] / nChs ) * length( aux04 ) ) / ( σ ) ;
end
Selection = sort( aux04 );
C1 = length( aux04 ) % σ == 0;
C2 = NRecFrames % σ == 0;
extras = setdiff( aux03, aux04 );
C3 = !isempty( extras );
if C1
    ChannelsIndexes = reshape( Selection, Int( length( Selection ) / σ ), σ );
    FramesIndexes = [ 1, NRecFrames ];
elseif C2
    ChannelsIndexes = Selection;
    FramesIndexes = collect( 0 : Int( NRecFrames / σ ) : NRecFrames );
else
    println( "Se necesita otro corte" )
end;
```

```julia
using HDF5
binlenght = Int( NRecFrames / σ );
BIN = Array{ UInt16 }( undef, length( ChannelsIndexes ), binlenght );
Raw = h5open( Variables[ "BRWname" ], "r" )[ Variables[ "Raw" ] ];
n = 1
for m = 1:length( ChannelsIndexes )
    BIN[ m, : ] .= Raw[ ChannelsIndexes[ m ] * n : nChs : nChs * binlenght * n ];
end
BIN = Digital2Analogue( Variables, BIN );
```
