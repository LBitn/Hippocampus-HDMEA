```julia
using Plots
```

```julia
EventsRAWStat = DetectedEvents[ 1 ]; EventsRAWDin = DetectedEvents[ 2 ];
EventsMUAStat = DetectedEvents[ 3 ]; EventsMUADin = DetectedEvents[ 4 ];
EventsLFPStat = DetectedEvents[ 5 ]; EventsLFPDin = DetectedEvents[ 6 ];
```
